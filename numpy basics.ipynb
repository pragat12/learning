#Iterating Over Array

```python
a = np.arange(0, 45, 5)
a = a.reshape(3,3)
a
```

```python
for x in np.nditer(a):
    print(x)
```

#Iteration order (c-style and f-style)

```python
print(a)
for x in np.nditer(a, order="C"):
    print(x)
for x in np.nditer(a, order = 'F'):
    print(x)
```

Joining Arrays

```python
a = np.array([[1,2],[3,4]])
print('first array')
print(a)
b = np.array([[5,6],[7,8]])
print('second array')
print(b)
print('\n')
print(np.concatenate((a,b)))
print('\n')
print('Joining the two arrays along axis 1:')
print(np.concatenate((a,b), axis = 1))
print('\n')
print('Joining the two arrays along axis 0:')
print(np.concatenate((a,b), axis = 0))
```

```python
x = np.arange(0,3 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
#Create a 6*6 two-dimensional array and 1 and 0 place alternatively across the diagonals

z = np.zeros((6,6),dtype=int)
z[1::2, ::2] = 1
z[::2, 1::2] = 1
z
```

```python
#find the total number and locations of missing values in the array
z = np.random.rand(10,10)
z[np.random.randint(10,size=5), np.random.randint(10, size = 5)] = np.nan
z
```

```python
print("Total number of missing value : \n", np.isnan(z).sum())
print("INdexes of missing vlaue : \n", np.argwhere(np.isnan(z)))
inds = np.where(np.isnan(z))
print(inds)
z[inds] = 0
print(z)
```
